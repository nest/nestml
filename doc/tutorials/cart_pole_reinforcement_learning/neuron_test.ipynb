```python
from pynestml.codegeneration.nest_code_generator_utils import NESTCodeGeneratorUtils

# generate and build code
input_layer_module_name, input_layer_neuron_model_name = \
   NESTCodeGeneratorUtils.generate_code_for("../../../models/neurons/ignore_and_fire_neuron.nestml")

# ignore_and_fire
output_layer_module_name, output_layer_neuron_model_name, output_layer_synapse_model_name = \
    NESTCodeGeneratorUtils.generate_code_for("iaf_psc_exp_neuron.nestml",
                                             "neuromodulated_stdp_synapse.nestml",
                                             post_ports=["post_spikes"],
                                             logging_level="DEBUG",
                                             codegen_opts={"delay_variable": {"neuromodulated_stdp_synapse": "d"},
                                                           "weight_variable": {"neuromodulated_stdp_synapse": "w"}})
```

```python
import numpy as np
import matplotlib.pyplot as plt

class Spiking_PlotRenderer():
    def __init__(self) -> None:

        plt.ion()
        
        self.fig, self.ax = plt.subplots(nrows=3)
        
        # Top plot for spikes
        self.ax[0].set_xlabel("Time [ms]")
        self.ax[0].set_title("Input Spikes")
        self.ax[0].set_xlim(0, 40)  # Set the x-axis limits
        
        # Middle plot for membrane potential
        self.ax[1].set_ylabel("V_m [mV]")
        self.ax[1].set_xlabel("Time [ms]")
        self.ax[1].set_xlim(0, 40)  # Set the x-axis limits
        self.ax[1].invert_yaxis()

        self.ax[2].set_ylabel("g_e [no unit]")
        self.ax[2].set_xlabel("Time [ms]")
        self.ax[2].set_xlim(0, 40)  # Set the x-axis limits
        self.ax[2].invert_yaxis()
    
        self.fig.show()
        
    def update(self, data) -> None:
        if data is None: return

        self.ax[0].cla()
        self.ax[1].cla()
        self.ax[2].cla()

        self.ax[0].set_ylabel("Input Spike")
        self.ax[0].plot(data["input_spikes"]["times"], data["input_spikes"]["senders"], ".k", markersize=1)
        self.ax[0].set_xlim(np.min(data["multimeter_output_events"]["times"])-5, np.max(data["multimeter_output_events"]["times"])+5)
        
        self.ax[1].set_ylabel("V_m [mV]")
        self.ax[1].plot(data["multimeter_output_events"]["times"], data["multimeter_output_events"]["V_m"], 'r')
        self.ax[1].set_xlim(np.min(data["multimeter_output_events"]["times"])-5, np.max(data["multimeter_output_events"]["times"])+5)

        self.ax[2].set_ylabel("g_e [no unit]")
        self.ax[2].set_xlabel("Time [ms]")
        self.ax[2].plot(data["multimeter_output_events"]["times"], data["multimeter_output_events"]["g_e"], 'b')
        self.ax[2].set_xlim(np.min(data["multimeter_output_events"]["times"])-5, np.max(data["multimeter_output_events"]["times"])+5)
        
        self.fig.canvas.draw()
        self.fig.canvas.flush_events()
        plt.pause(0.0001)
```

```python
import nest
%matplotlib qt
nest.ResetKernel()
nest.Install(input_layer_module_name)   # makes the generated NESTML model available
nest.Install(output_layer_module_name)   # makes the generated NESTML model available

input_neuron = nest.Create(input_layer_neuron_model_name)
output_neuron = nest.Create(output_layer_neuron_model_name)

nest.Connect(input_neuron, output_neuron, syn_spec={"synapse_model": output_layer_synapse_model_name,
                                                    "w": 0.1 + nest.random.uniform(0., 0.02)})

spike_recorder_input = nest.Create("spike_recorder")
nest.Connect(input_neuron,spike_recorder_input)

multimeter_output= nest.Create('multimeter', 1, {'record_from': ['V_m', 'g_e'], 'interval': 0.1})
nest.Connect(multimeter_output ,output_neuron)

input_neuron.firing_rate = 25.

plotter = Spiking_PlotRenderer()
for _ in range(10000):
        nest.Simulate(10)
        
        #passed onto Spiking_Plot_Renderer()
        plot_data = {
            "input_spikes": nest.GetStatus(spike_recorder_input, keys="events")[0],
            "multimeter_output_events": multimeter_output.get("events"),
        }
        plotter.update(plot_data)
        spike_recorder_input.n_events = 0
        multimeter_output.n_events = 0
```
